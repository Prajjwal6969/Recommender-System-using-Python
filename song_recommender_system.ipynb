Importing all dependencies

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
import difflib
from sklearn.metrics.pairwise import cosine_similarity
```

importing our dataset

```python
song_data = pd.read_csv('song_data.csv')
print(song_data.head())
song_data.info()
```

Select Main Features

```python
selected_features = ['Song Title','Hindi Lyrics','English Translations']
print(selected_features)
```

```python
combined_features = song_data['Song Title'] + ' '+song_data['Hindi Lyrics']+' '+song_data['English Translations']
print(combined_features)
```

Change Data into numerical features

```python
vectorizer = TfidfVectorizer()
numerical_feature = vectorizer.fit_transform(combined_features)
print(numerical_feature)
```

Check the simialrity Scores

```python
similarity = cosine_similarity(numerical_feature)
print(similarity)
```

Main Recommender system

```python
song_name = input('Enter your favourite song name')
list_of_all_titles = song_data['Song Title'].tolist()
find_close_match = difflib.get_close_matches(song_name,list_of_all_titles)
print(find_close_match)
close_match = find_close_match[0]
print(close_match)
index_of_song = song_data[song_data['Song Title'] == close_match].index[0]
print(index_of_song)
similarity_score = list(enumerate(similarity[index_of_song]))
print(similarity_score)
sorted_similar_score = sorted(similarity_score,key = lambda x:x[1],reverse=True)
print(sorted_similar_score)

print("songs suggested for you are: ")
i = 1

for songs in sorted_similar_score:
  index = songs[0]
  title_of_songs = song_data[song_data.index == index]['Song Title'].values[0]
  if(i<10):
    print(i,title_of_songs)
    i = i+1
```
